### Ejercicios de apropiación y aplicación

1. Gráfica las siguientes funciones trigonometricas 

+ $f(t)=\sin(4t)$

+ $f(t)=\sin(4t)+\cos(5t)$

2. Muestra si los siguientes enunciados son identidades trigonométricas 

+ $\sin(x)\sin(3x) = \frac{1}{2}[\cos(2x) - \cos(4x)]$
 
+ $\cos^4(x) - \sin^4(x) = \cos(2x)$

+  $\frac{1 + \cot^2(x)}{1 + \tan^2(x)} = \frac{\cos^2(x)}{\sin^2(x)}$

3. Resuelve las siguientes ecuaciones trigonométricas en el intervalo $ 0 \leq x \leq 2\pi $:

+ $2\cos(2x) + 1 = 0$

+ $2\sin^2(x) - 3\sin(x) + 1 = 0$

+ $\tan(3x) = 1$

3. Soluciones los siguientes triángulos rectángulos:

-  En un triángulo rectángulo, conocemos que uno de sus catetos mide 15 metros, y el ángulo opuesto a este cateto es de $50^o$. Calcula la hipotenusa, el otro cateto, el ángulo faltante y el área del triángulo.

 
- Desde un punto $P$ en el suelo, se observa la parte superior de una antena con un ángulo de elevación de $18^o$. Si avanzamos 15 metros hacia la antena, el ángulo de elevación aumenta a $40^o$. Calcula la altura de la antena.
