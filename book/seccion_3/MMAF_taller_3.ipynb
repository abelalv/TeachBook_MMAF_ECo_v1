### A.12 Trayectoria de un proyectil  
Un proyectil es lanzado hacia arriba desde el suelo. La altura $h(t)$ en metros del proyectil después de $t$ segundos está dada por la fórmula:  
$$
h(t) = 20t - 5t^2
$$
- (a) ¿En qué intervalos de tiempo el proyectil estará a más de $15$ metros de altura?  
  (Sugerencia: Resuelva la inecuación $20t - 5t^2 > 15$).  
- (b) ¿En qué instante exacto el proyectil alcanza los $15$ metros de altura?  
  (Sugerencia: Resuelva la ecuación cuadrática $20t - 5t^2 = 15$).
