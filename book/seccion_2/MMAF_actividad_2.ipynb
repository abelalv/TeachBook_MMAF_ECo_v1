## Una breve historia del contenedor martimo

El contenedor martimo, tambin llamado "container" (trmino ingls), se remonta a la Segunda Guerra Mundial. Era utilizado para el transporte seguro de materiales blicos.

Fue creado por el estadounidense Malcolm McLean, un camionero que naci el 1913 en una familia de clase media de Carolina del Norte. Fue en 1953, cuando este empresario observ que casi todos los puertos de la costa oeste estaban completamente saturados. Para optimizar el tiempo decidi cargar directamente los remolques de sus camiones a los barcos.

Partiendo de este modelo operativo tambin se percat que supona un ahorro de tiempo, pero no era demasiado rentable, ya que el espacio utilizado por los remolques era demasiado grande. A partir de ah, le surge la idea de quitar el chasis y embarcar solo la parte del remolque, es decir, la "caja" en s. Y as fue como invent el contenedor que conocemos hoy en da.

En menos de diez aos, el uso de los contenedores de transporte se extendi por todo el mundo, ya que permiti optimizar el espacio a bordo y ahorrar tiempo para las operaciones de manipulacin. Fue en 1961 cuando los containers de McLean se modifican para implementar una estandarizacin internacional de los contenedores y realizar las ISO, 20 ', 30' y 40 como dimensiones estndares.

Hoy en da, estn hechos mayoritariamente de acero, aunque tambin podemos encontrarlos de aluminio. Adems, existe una gran variedad de contenedores para compensar la escasez y optimizar el transporte de mercancas.





Es importante saber que los contenedores cumplen con las normas internacionales ISO que rigen el transporte de mercancas.  
Dimensiones externas de los contenedores (mm)

	Dry 40 pies: 12 192 mm (longitud), 2 438 mm (ancho) y 2 591 mm (altura).

	High Cube 40 pies: 12 192 mm (longitud), 2 438 mm (ancho) y 2 698 mm (altura).
