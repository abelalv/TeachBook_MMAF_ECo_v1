**Nota previa:**  
En una función cuadrática de la forma  
$$
f(x) = ax^2 + bx + c,
$$  
el **vértice** es el punto donde la parábola alcanza su **valor máximo** (si $a<0$) o su **valor mínimo** (si $a>0$).  

- La coordenada en $x$ del vértice se calcula con  
  $$
  x_v = -\frac{b}{2a}.
  $$  
- El valor correspondiente en $y$ se obtiene evaluando $f(x_v)$.
